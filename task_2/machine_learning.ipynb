```python
def encode(dataframe, columns):
    label_encoder = LabelEncoder()

    for column in columns:
        if column in dataframe.columns:
            mapped_values = label_encoder.fit_transform(dataframe[column])
            dataframe[f"{column}_encoded"] = mapped_values
        else:
            print(f"Column '{column}' not found in the DataFrame.")

    return dataframe
```

```python
# Feature engineering for easier training
encoded_columns = encode(df,['sales_channel','trip_type','route','flight_day','booking_origin'])
training_df = encoded_columns.drop(['sales_channel','trip_type','flight_day','route','booking_origin'], axis = 1)
```

```python
# let's find the correlation

sns.heatmap(training_df.corr(),annot=True)
```

```python
# 'booking_complete' column is target variable

X = training_df.drop('booking_complete', axis = 1)
y= training_df['booking_complete']
```

```python
# Split the data into training and testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2, random_state=42)
```

```python
# Feature Scaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
# Train the model

model = LogisticRegression()
model.fit(X_train,y_train)
```

```python
# Make predictions

y_pred = model.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
confusion_mat = confusion_matrix(y_test, y_pred)
classification_rep = classification_report(y_test, y_pred)

print(f"Accuracy: {accuracy}")
print(f"Confusion Matrix:\n{confusion_mat}")
print(f"Classification Report:\n{classification_rep}")
```

```python
# grab the coefficients and names 
coefficients = model.coef_[0]
feature_names = X.columns

# Create a DataFrame to store feature names and their corresponding coefficients
feature_coefficients = pd.DataFrame({'Feature': feature_names, 'Coefficient': coefficients})
print(feature_coefficients)
```

```python
# grab the coefficients and names 
coefficients = model.coef_[0]
feature_names = X.columns

# Create a DataFrame to store feature names and their corresponding coefficients
feature_coefficients = pd.DataFrame({'Feature': feature_names, 'Coefficient': coefficients})

# Sort the DataFrame by absolute coefficient values for better visualization
feature_coefficients['Abs_Coefficient'] = np.abs(feature_coefficients['Coefficient'])
sorted_features = feature_coefficients.sort_values(by='Abs_Coefficient', ascending=False)

# Plot the coefficients
plt.figure(figsize=(10, 6))
plt.barh(sorted_features['Feature'], sorted_features['Coefficient'], color='skyblue')
plt.xlabel('Coefficient Value')
plt.title('Feature Coefficients')
plt.grid(axis='x', linestyle='--', alpha=0.6)
plt.show()
```
